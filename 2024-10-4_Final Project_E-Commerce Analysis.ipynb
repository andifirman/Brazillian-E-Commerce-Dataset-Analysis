```python
# Menghitung jumlah review_comment_title yang kosong berdasarkan skor review

order_reviews_df.groupby(by="review_score").agg({
    "review_comment_title": lambda x: x.isnull().sum()
})
```

### 2.4 Explore Korelasi Antar Tabel

#### 2.4.1 Tabel customers_df dan Tabel orders_df

Selanjutnya kita akan mengeskplorasi korelasi dari dua buah tabel yaitu antara tabel customers_df dan tabel orders_df. Untuk menggabungkan dua buah tabel dapat digunakan konsep **LEFT JOIN** dengan memanfaatkan fungsi merge. Gunakan customer_id yang merupakan primary key dari tabel customers_df dan tabel orders_df. 

```python
customers_orders_df = pd.merge(
    left=customers_df,
    right=orders_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)

customers_orders_df.head()
```

#### 2.4.2 Tabel payments_df dan Tabel reviews_df

Selanjutnya kita akan mengeskplorasi korelasi dari dua buah tabel yaitu antara tabel order_payments_df dan tabel order_reviews_df. Untuk menggabungkan dua buah tabel dapat digunakan konsep **LEFT JOIN** dengan memanfaatkan fungsi merge. Gunakan order_id yang merupakan primary key dari tabel order_payments_df dan tabel order_reviews_df. 

```python
payments_review_df = pd.merge(
    left=order_payments_df,
    right=order_reviews_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)

payments_review_df.head()
```

```python
payments_review_df.sort_values(by="payment_value", ascending=False).head()
```

```python
payments_review_df.groupby(by="payment_type").agg({
    "order_id": "nunique",
    "payment_value": ["min", "max"]
})
```

#### 2.4.3 Tabel customers_orders_df dan Tabel payments_df

Masih dengan menggunakan **LEFT JOIN**, selanjutnya kita akan menggabungkan tabel yang sebelumnya telah digabung (tabel customers_orders_df) dengan tabel order_payments_df dengan menggunakan atribut order_id sebagai primary key dari kedua tabel tersebut. 

```python
merged_customers_df = pd.merge(
    left=customers_orders_df,
    right=order_payments_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)

merged_customers_df.head()
```

```python
items_sellers_df = pd.merge(
    left=order_items_df,
    right=sellers_df,
    how="left",
    left_on="seller_id",
    right_on="seller_id"
)

items_sellers_df.head()
```
